### <span style="color:lightblue; font-weight:bold;">Python Mini Project - ADDA 24 (A Fun Game)</span>


This a fun and interactive Python-based mini cricket game that allows two players to compete in a virtual cricket match with multiple overs options (1, 3, or 5 overs). The game starts with a coin toss to decide which player will bat or bowl first. Players take turns to bat and bowl, with each ball's outcome being determined by random scoring and the player's input to predict the opponent's score.

```python
import mysql.connector as sql
import random as rd

obj = sql.connect(host="localhost", user="root", password="123456", database="cricket_project")

scorelstp1=[]
scorelstp2=[]
scorelstp13over=[]
scorelstp23over=[]
scorelstp15over=[]
scorelstp25over=[]
matchCount=[0,0,0]
playerWin=[0,0,0]
scoreLists=[[scorelstp1, scorelstp2], [scorelstp13over, scorelstp23over], [scorelstp15over, scorelstp25over]]

def game(overs):
    print("*" * 20, f"It's a {overs} overs game", "*" * 20)
    toss()
    totalScores = [0, 0]
    for i in range(2):
        print("*" * 80)
        print()
        print(f"Player {i+1}'s turn to bat")
        print("-" * 80)
        print(f'First Innings {"Ends" if i==1 else "Starts"}')
        print("-" * 80)
        for _ in range(1, (overs * 6) + 1):
            print("-" * 80)
            print()
            score = rd.randint(1, 6)
            out = int(input("Enter Player Out Number (Should be 1-6): "))
            print("-" * 80)
            print()
            print("Your this ball score:", score)
            if out == score:
                print("Oops, you got out!")
                print(f"Player {i+1}'s final score is", totalScores[i])
                break
            else:
                totalScores[i] += score
                print("-" * 80)
                print(f"Player {i+1}'s Total score is", totalScores[i])
                print("-" * 80)
        print("Final Score", totalScores[i])
        print("*" * 80)
        print()
    update_score(matchCount[overs//2], playerWin[overs//2], sum(scoreLists[overs//2][0]), totalScores[0], totalScores[1], obj, overs)

def toss():
    try:
        HT = ["head", "tail"]
        batAndBowl = ['bat', 'ball']
        print("Player 1 will call")
        T = input("Please Choose Head or Tail: ").lower()
        if T not in HT:
            raise ValueError()
        ht = rd.choice(HT)
        player = HT.index(ht) + 1
        print(f"Player {player} won the toss")
        bat = input("Select Bat or Ball First: ").lower()
        if bat == batAndBowl[0]:
            print(f"Player {player} elected to bat first")
        elif bat == batAndBowl[1]:
            print(f"Player {player} elected to bowl")
    except:
        print("Please make a right choice")
        toss()

def update_score(match_played, match_won, total_score, highest_score, lowest_score, obj, overs):
    try:
        average_score = total_score / match_played
    except ZeroDivisionError:
        average_score = 0
    cobj = obj.cursor()
    c = cobj.execute(
        f"UPDATE {overs}overmatch SET match_played={match_played}, match_won={match_won}, Total_score={total_score}, Average_score={average_score}, highest_score={highest_score}, lowest_score={lowest_score} where player=1")
    d = cobj.execute(
        f"UPDATE {overs}overmatch SET match_played={match_played}, match_won={match_played-match_won}, Total_score={total_score}, Average_score={average_score}, highest_score={highest_score}, lowest_score={lowest_score} where player=2")
    obj.commit()

def matchover():
    try:
        while True:
            print("""Please enter no. of overs
                Choice Available 1 overs match -- 3 overs match -- 5 overs match or 'end' to Finish""")
            mo = input("Enter no. of over : ").lower()
            if mo in ["1", "3", "5"]:
                matchCount[int(mo)//2] += 1
                game(int(mo))
            elif mo == "end":
                print("Thanks for Playing have a nice day")
                break
            else:
                print("-" * 80)
                print("Please Enter the right choice")
                print("-" * 80)
                print()
    except:
        print("Please make a right Choice")

matchover()

scorelstp1, scorelstp2 = [], []
scorelstp13over, scorelstp23over = [], []
scorelstp15over, scorelstp25over = [], []
matchCount1, matchCount3, matchCount5 = 0, 0, 0
player1win1, player2win1 = 0, 0
player1win3, player2win3 = 0, 0
player1win5, player2win5 = 0, 0

def game(overs):
    print("*" * 20, f"It's a {overs} overs game", "*" * 20)
    toss()
    totalScore1, totalScore2 = 0, 0
    for i in range(2):
        print("*" * 80)
        print()
        print(f"Player {i+1}'s turn to bat")
        print("-" * 80)
        print(f'First Innings {"Ends" if i==1 else "Starts"}')
        print("-" * 80)
        totalScore = 0
        for _ in range(1, (overs * 6) + 1):
            print("-" * 80)
            print()
            score = rd.randint(1, 6)
            out = int(input("Enter Player Out Number (Should be 1-6): "))
            print("-" * 80)
            print()
            print("Your this ball score:", score)
            if out == score:
                print("Oops, you got out!")
                print(f"Player {i+1}'s final score is", totalScore)
                break
            else:
                totalScore += score
                print("-" * 80)
                print(f"Player {i+1}'s Total score is", totalScore)
                print("-" * 80)
        if i == 0:
            totalScore1 = totalScore
        else:
            totalScore2 = totalScore
        print("Final Score", totalScore)
        print("*" * 80)
        print()
    update_score(overs, totalScore1, totalScore2)

def toss():
    try:
        HT = ["head", "tail"]
        batAndBowl = ['bat', 'ball']
        print("Player 1 will call")
        T = input("Please Choose Head or Tail: ").lower()
        if T not in HT:
            raise ValueError()
        ht = rd.choice(HT)
        player = HT.index(ht) + 1
        print(f"Player {player} won the toss")
        bat = input("Select Bat or Ball First: ").lower()
        if bat == batAndBowl[0]:
            print(f"Player {player} elected to bat first")
        elif bat == batAndBowl[1]:
            print(f"Player {player} elected to bowl")
    except:
        print("Please make a right choice")
        toss()

def update_score(overs, totalScore1, totalScore2):
    match_played = 0
    match_won1 = 0
    total_score1 = 0
    average_score1 = 0
    highest_score1 = 0
    lowest_score1 = 0
    match_won2 = 0
    total_score2 = 0
    average_score2 = 0
    highest_score2 = 0
    lowest_score2 = 0
    
    match_played = 1 if overs == 1 else 3 if overs == 3 else 5
    if totalScore1 > totalScore2:
        match_won1 = 1
    elif totalScore2 > totalScore1:
        match_won2 = 1
    total_score1 = totalScore1
    average_score1 = total_score1 / match_played if match_played > 0 else 0
    highest_score1 = totalScore1
    lowest_score1 = totalScore1
    total_score2 = totalScore2
    average_score2 = total_score2 / match_played if match_played > 0 else 0
    highest_score2 = totalScore2
    lowest_score2 = totalScore2
    
    cobj = obj.cursor()
    c = cobj.execute(
        f"UPDATE {overs}overmatch SET match_played={match_played}, match_won={match_won1}, Total_score={total_score1}, Average_score={average_score1}, highest_score={highest_score1}, lowest_score={lowest_score1} where player=1")
    d = cobj.execute(
        f"UPDATE {overs}overmatch SET match_played={match_played}, match_won={match_won2}, Total_score={total_score2}, Average_score={average_score2}, highest_score={highest_score2}, lowest_score={lowest_score2} where player=2")
    obj.commit()

def matchover():
    try:
        while True:
            print("""Please enter no. of overs
                Choice Available 1 overs match -- 3 overs match -- 5 overs match or 'end' to Finish""")
            mo = input("Enter no. of over : ").lower()
            if mo == "1":
                game(1)
            elif mo == "3":
                game(3)
            elif mo == "5":
                game(5)
            elif mo == "end":
                print("Thanks for Playing have a nice day")
                break
            else:
                print("-" * 80)
                print("Please Enter the right choice")
                print("-" * 80)
                print()
    except:
        print("Please make a right Choice")

matchover()
```
